# Dataset

```python
from datasets import load_dataset

# Muat dataset dari Hugging Face
dataset = load_dataset("shobrunjb/skincare-product-review-indo-v1")

# Konversi dataset menjadi DataFrame Pandas
df = dataset['train'].to_pandas()

# Tampilkan beberapa baris pertama
df.head()
```

```python
# Pastikan kolom 'stem_reviews' dalam bentuk string
df['no_whitespace_reviews'] = df['no_whitespace_reviews'].astype(str)
```

# Modeling

```python
from transformers import BertTokenizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Inisialisasi tokenizer IndoBERT
tokenizer = BertTokenizer.from_pretrained('indobenchmark/indobert-base-p1')

# Tokenisasi teks
def tokenize_function(text):
    return tokenizer(text, padding='max_length', truncation=True, max_length=128, return_tensors="pt")

df['input_ids'] = df['no_whitespace_reviews'].apply(lambda x: tokenize_function(x)['input_ids'].squeeze())
df['attention_mask'] = df['no_whitespace_reviews'].apply(lambda x: tokenize_function(x)['attention_mask'].squeeze())

# Encoding label
label_encoder_task1 = LabelEncoder()
df['label_encoded'] = label_encoder_task1.fit_transform(df['label'])

label_encoder_task2 = LabelEncoder()
df['sentimen_encoded'] = label_encoder_task2.fit_transform(df['sentimen'])

# Split dataset menjadi training dan test set untuk kedua task
X = df[['input_ids', 'attention_mask']]
y_task1 = df['label_encoded']
y_task2 = df['sentimen_encoded']

X_train, X_test, y_train_task1, y_test_task1 = train_test_split(X, y_task1, test_size=0.2, random_state=42)
_, _, y_train_task2, y_test_task2 = train_test_split(X, y_task2, test_size=0.2, random_state=42)
```

```python
import torch
from transformers import BertModel
from torch import nn
from torch.utils.data import DataLoader, TensorDataset

# Definisikan model dengan Dropout
class IndoBERTMultiTaskClassifier(nn.Module):
    def __init__(self, bert_model_name, num_labels_task1, num_labels_task2, dropout_rate=0.3):
        super(IndoBERTMultiTaskClassifier, self).__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        self.dropout = nn.Dropout(dropout_rate)
        self.classifier_task1 = nn.Linear(self.bert.config.hidden_size, num_labels_task1)
        self.classifier_task2 = nn.Linear(self.bert.config.hidden_size, num_labels_task2)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids=input_ids, attention_mask=attention_mask)
        cls_output = outputs[1]  # CLS token
        cls_output = self.dropout(cls_output)

        logits_task1 = self.classifier_task1(cls_output)
        logits_task2 = self.classifier_task2(cls_output)

        return logits_task1, logits_task2
```

```python
from torch.utils.data import DataLoader, TensorDataset


# DataLoader untuk training dan testing dengan kedua task
train_dataset = TensorDataset(torch.stack(X_train['input_ids'].tolist()),
                              torch.stack(X_train['attention_mask'].tolist()),
                              torch.tensor(y_train_task1.values),
                              torch.tensor(y_train_task2.values))

test_dataset = TensorDataset(torch.stack(X_test['input_ids'].tolist()),
                             torch.stack(X_test['attention_mask'].tolist()),
                             torch.tensor(y_test_task1.values),
                             torch.tensor(y_test_task2.values))

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
# Inisialisasi model multitask
# Inisialisasi model multitask
model = IndoBERTMultiTaskClassifier(
    'indobenchmark/indobert-base-p1',
    num_labels_task1=len(label_encoder_task1.classes_),
    num_labels_task2=len(label_encoder_task2.classes_)
)

# Tentukan loss function dan optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=2e-5)

# Inisialisasi parameter untuk early stopping
patience = 2  # Jumlah epoch yang boleh dilewati tanpa perbaikan
best_loss = np.inf
epochs_no_improve = 0

# Modifikasi fungsi pelatihan untuk multi-task learning
def train_model_with_early_stopping_multitask(model, train_loader, test_loader, criterion, optimizer, epochs=10, patience=2):
    global best_loss
    global epochs_no_improve

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for input_ids, attention_mask, labels_task1, labels_task2 in train_loader:
            optimizer.zero_grad()
            outputs_task1, outputs_task2 = model(input_ids, attention_mask)
            loss_task1 = criterion(outputs_task1, labels_task1)
            loss_task2 = criterion(outputs_task2, labels_task2)
            loss = loss_task1 + loss_task2
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        avg_train_loss = total_loss / len(train_loader)

        # Evaluasi pada test set
        model.eval()
        total_test_loss = 0
        with torch.no_grad():
            for input_ids, attention_mask, labels_task1, labels_task2 in test_loader:
                outputs_task1, outputs_task2 = model(input_ids, attention_mask)
                loss_task1 = criterion(outputs_task1, labels_task1)
                loss_task2 = criterion(outputs_task2, labels_task2)
                total_test_loss += (loss_task1 + loss_task2).item()

        avg_test_loss = total_test_loss / len(test_loader)

        print(f"Epoch {epoch+1}/{epochs} - Train Loss: {avg_train_loss:.4f} - Test Loss: {avg_test_loss:.4f}")

        # Cek untuk early stopping
        if avg_test_loss < best_loss:
            best_loss = avg_test_loss
            epochs_no_improve = 0
            best_model = model.state_dict()  # Simpan model terbaik
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping triggered!")
                model.load_state_dict(best_model)  # Load model terbaik sebelum overfitting
                break

train_model_with_early_stopping_multitask(model, train_loader, test_loader, criterion, optimizer)
```

```python
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

def evaluate_model_multitask(model, test_loader):
    model.eval()
    predictions_task1, true_labels_task1 = [], []
    predictions_task2, true_labels_task2 = [], []

    with torch.no_grad():
        for input_ids, attention_mask, labels_task1, labels_task2 in test_loader:
            outputs_task1, outputs_task2 = model(input_ids, attention_mask)
            logits_task1 = outputs_task1.detach().cpu().numpy()
            logits_task2 = outputs_task2.detach().cpu().numpy()

            pred_labels_task1 = np.argmax(logits_task1, axis=1)
            pred_labels_task2 = np.argmax(logits_task2, axis=1)

            predictions_task1.extend(pred_labels_task1)
            predictions_task2.extend(pred_labels_task2)

            true_labels_task1.extend(labels_task1.cpu().numpy())
            true_labels_task2.extend(labels_task2.cpu().numpy())

    print("Task 1 - Deteksi Fake Review")
    print(classification_report(true_labels_task1, predictions_task1, target_names=label_encoder_task1.classes_))

    print("Task 2 - Klasifikasi Sentimen")
    print(classification_report(true_labels_task2, predictions_task2, target_names=label_encoder_task2.classes_))

    accuracy_task1 = accuracy_score(true_labels_task1, predictions_task1)
    accuracy_task2 = accuracy_score(true_labels_task2, predictions_task2)

    print(f"Task 1 Accuracy: {accuracy_task1 * 100:.2f}%")
    print(f"Task 2 Accuracy: {accuracy_task2 * 100:.2f}%")

    return true_labels_task1, predictions_task1, true_labels_task2, predictions_task2

# Panggil fungsi evaluasi untuk mendapatkan hasil
true_labels_task1, predictions_task1, true_labels_task2, predictions_task2 = evaluate_model_multitask(model, test_loader)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()

def create_confusion_matrices(true_labels_task1, predictions_task1, true_labels_task2, predictions_task2):
    # Confusion Matrix untuk Task 1 (Deteksi Fake Review)
    cm_task1 = confusion_matrix(true_labels_task1, predictions_task1)
    plot_confusion_matrix(cm_task1, classes=label_encoder_task1.classes_, title='Confusion Matrix - Task 1 (Fake Review Detection)')

    # Confusion Matrix untuk Task 2 (Klasifikasi Sentimen)
    cm_task2 = confusion_matrix(true_labels_task2, predictions_task2)
    plot_confusion_matrix(cm_task2, classes=label_encoder_task2.classes_, title='Confusion Matrix - Task 2 (Sentiment Classification)')

# Panggil fungsi untuk membuat confusion matrix
create_confusion_matrices(true_labels_task1, predictions_task1, true_labels_task2, predictions_task2)
```
